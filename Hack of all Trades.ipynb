```python
def Stacking(model,train,y,test,n_fold):
    folds=KFold(n_splits=n_fold,random_state=42)
    test_pred=np.zeros((test.shape[0],),float)
    #test_pred=np.empty((0,1),float)
    train_pred=np.empty((0,1),float)
    #train_pred = np.zeros((train.shape[0],))
    for train_indices,val_indices in folds.split(train,y.values):
        #print(train_indices)
        #print(val_indices)
        x_train,x_val=train.iloc[train_indices],train.iloc[val_indices]
        y_train,y_val=y.iloc[train_indices],y.iloc[val_indices]
        model.fit(X=x_train,y=y_train)
        train_pred=np.append(train_pred,model.predict(x_val))
        #train_pred[val_indices]=model.predict(x_val)
        test_pred=np.column_stack((test_pred,model.predict(test)))
    return np.mean(test_pred[:,1:],axis=1),train_pred
```
